# Обучение и оптимизация модели из класса DecisionTreeRegressor

Разделяем исходную выборку на тренировочную и валидационную. Определяем оптимальную глубину дерева. Каким образом? Выбираем ту глубину дерева, для которой будет наименьшая MAE (средняя абсолютная ошибка) на валидационной выборке.


```python
import pandas as pd
```

```python
melbourne_file_path = 'https://raw.githubusercontent.com/juliana-zh/machine_learning/main/datasets/melb_data.csv'
# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.describe()
```

```python
# remove missing objects (or columns)
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
# select prediction target
y = melbourne_data.Price
```

```python
melbourne_data.columns
```

```python
# select subset of features
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
# Fill in argument to make optimal size
final_model = DecisionTreeRegressor(max_leaf_nodes=500, random_state=0)

# fit the final model
final_model.fit(X, y)
```
